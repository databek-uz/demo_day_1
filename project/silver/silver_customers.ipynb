```python
import pyspark.sql.functions as F
```

```python
bronze_df = spark.table("workspace.bronze.customers")
```

```python
silver_df = bronze_df.withColumnsRenamed({
    "cst_id": "customer_id",
    "cst_key": "customer_key",
    "cst_firstname": "first_name",
    "cst_lastname": "last_name",
    "cst_marital_status": "marital_status",
    "cst_gndr": "gender",
    "cst_create_date": "create_date"
}) \
 .withColumn("gender", F.when(F.col("gender") == "M", "Male").otherwise("Female")) \
 .withColumn("marital_status", F.when(F.col("marital_status") == "S", "Single").otherwise("Married"))
```

```python
silver_df.write.mode("overwrite").saveAsTable("silver.dim_customers")

# TODO: we need implement merge/incermental logic.
```
