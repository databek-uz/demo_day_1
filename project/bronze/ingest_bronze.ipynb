```python
import pyspark.sql.functions as F
```

```python
dbutils.widgets.text("raw_file", "", "Raw File Path")
dbutils.widgets.text("bronze_table", "", "Bronze Table Path")

raw_file = dbutils.widgets.get("raw_file")
bronze_table = dbutils.widgets.get("bronze_table")
```

```python
bronze_df = spark.read.format("csv").option("header", "true").load(raw_file, inferSchema=True)
bronze_df.write.mode("overwrite").saveAsTable(bronze_table)

# TODO: we need implement merge/incermental logic.
```
